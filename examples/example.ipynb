## Outline structure

### Headings become nodes

- `#` (H1) defines the mind map **root**.
- `##`, `###`, and deeper headings nest underneath.
- Non-heading cells belong to the nearest heading above them.
- Content before the first H1 is ignored in the map.

```python
# This code cell sits under "Headings become nodes"
print("Hello from Lumen")
```

### Drag to reorder

Use the grip on the left edge of a node to move it **before**, **inside**, or **after** another topic. The notebook cell order updates to match.
