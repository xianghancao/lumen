## Outline structure

### Headings become nodes

- `#` (H1) defines the mind map **root**.
- `##`, `###`, and deeper headings nest underneath.
- Non-heading cells belong to the nearest heading above them.
- Content before the first H1 is ignored in the map.

````python
# This code cell sits under "Headings become nodes"
print("Hello from Lumen")
````

### Drag to reorder

Use the grip on the left edge of a node to move it **before**, **inside**, or **after** another topic. The notebook cell order updates to match.

## Format toolbar

Double-click a markdown cell (or press **F2**) to enter edit mode. The format toolbar appears on the right side of the Lumen header.

| Control | What to try |
|---------|-------------|
| **Title** | Set outline heading H1–H6 (changes the mind map tree) |
| **Aa** | Bold, italic, underline, strikethrough, code, highlight, block quote |
| **List** | Bulleted, dashed, numbered, check list |
| **Table** | Grid picker — insert a Markdown table |
| **Image** | Markdown or HTML image syntax |
| **Link** | Markdown or HTML link syntax |
| **Font color** | Preset swatches or a custom color |

> **Title** changes structure. **Aa** styles affect content only.

## Empty nodes

Select any node in Lumen, then:

- Press **Tab** to insert a **child** node (blank markdown cell, correct tree level).
- Press **Enter** to insert a **sibling** node.

New nodes start **empty** — no `#` heading prefix and no placeholder text. Tree level is stored in cell metadata until you add your own markdown (including `#` headings if you want).
